```python
import math
import itertools
from geopy.distance import geodesic

# Constants
CO2_EMISSION_PER_KM = 0.2  # CO2 emission per kilometer traveled

# List of delivery points (latitude, longitude)
# delivery_points = [
#     (40.712776, -74.005974),  # Example point 1
#     (34.052235, -118.243683),  # Example point 2
#     (41.878113, -87.629799),  # Example point 3
#     (29.760427, -95.369804),  # Example point 4
#     (51.507351, -0.127758)  # Example point 5
# ]

delivery_points = [
    (48.1171, -1.6778),
    (48.1044, -1.6794),
    (48.1145, -1.6744),
    (48.1107, -1.6736),
    (48.1048, -1.6772)
]


# Calculate distances between all pairs of delivery points
distances = {}
for i in range(len(delivery_points)):
    for j in range(i + 1, len(delivery_points)):
        point1 = delivery_points[i]
        point2 = delivery_points[j]
        distance = geodesic(point1, point2).kilometers
        distances[(i, j)] = distance
        distances[(j, i)] = distance

# Calculate CO2 emissions for a given route
def calculate_emission(route):
    total_emission = 0.0
    for i in range(len(route) - 1):
        point1 = route[i]
        point2 = route[i + 1]
        distance = distances[(point1, point2)]
        emission = distance * CO2_EMISSION_PER_KM
        total_emission += emission
    return total_emission

# Generate all permutations of the delivery points
permutations = list(itertools.permutations(range(len(delivery_points))))

# Calculate emissions for all permutations
emissions = [calculate_emission(p) for p in permutations]

# Find the minimum and maximum emissions
min_emission = min(emissions)
max_emission = max(emissions)

# Find the routes corresponding to the minimum and maximum emissions
min_route = permutations[emissions.index(min_emission)]
max_route = permutations[emissions.index(max_emission)]

# Print the results
print("Minimum CO2 Emission:", min_emission, "kg")
print("Optimal Route:", min_route)
print()
print("Maximum CO2 Emission:", max_emission, "kg")
print("Non-Optimal Route:", max_route)
```

```python
print(math.ceil((max_emission / min_emission)*100), "%")
```

```python
import folium

# Create a map centered at the first delivery point
map_center = delivery_points[0]
m = folium.Map(location=map_center, zoom_start=12)

# Add markers for each delivery point
for i, point in enumerate(delivery_points):
    folium.Marker(
        location=point,
        popup=f"Point {i+1}",
        icon=folium.Icon(icon="cloud")
    ).add_to(m)

# Draw the optimal path in blue
optimal_path = [delivery_points[i] for i in min_route]
folium.PolyLine(
    optimal_path,
    color='blue',
    weight=3,
    popup=f"Optimal Path\nTotal CO2 Emission: {min_emission} kg"
).add_to(m)

# Create a new map for the non-optimal path
m2 = folium.Map(location=map_center, zoom_start=12)

# Add markers for each delivery point
for i, point in enumerate(delivery_points):
    folium.Marker(
        location=point,
        popup=f"Point {i+1}",
        icon=folium.Icon(icon="cloud")
    ).add_to(m2)

# Draw the non-optimal path in red
non_optimal_path = [delivery_points[i] for i in max_route]
folium.PolyLine(
    non_optimal_path,
    color='red',
    weight=3,
    popup=f"Non-Optimal Path\nTotal CO2 Emission: {max_emission} kg"
).add_to(m2)

# Display the maps
m
```
